```python
def get_input(observation):
    slot = observation['job_slot']
    max_len = np.array([20, 20], dtype=np.float32)
    m = (observation['machine'] / max_len)
    durations = dict()
    job_reqs = dict()
    #m *= 1.5
    idx = 0
    for l, r in zip(slot['lengths'], slot['resource_vectors']):
        if l is None:
            continue
        ret = np.zeros(shape=(max_job_length,), dtype=float)
        ret[l - 1] = 1.0
        p = np.concatenate([r / max_len, ret, [0]])
        input_matrix[0, idx, : ] = p
        durations[idx] = l
        job_reqs[idx] = torch.FloatTensor(r / max_len)
        idx += 1
    p = np.zeros_like(input_matrix[0, idx, :])
    p[-1] = 1
    input_matrix[0, idx, :] = p

    return m, input_matrix[:, :(1 + idx), :], durations, job_reqs
```
